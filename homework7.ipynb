```python
from collections import Counter
def solve():
    try:
        arr = list(map(int, input().split()))
    except ValueError:
        print("Vui long chi nhap so nguyen.")
        return
    counts = Counter(arr)
    y = 0
    for x, f in counts.items():
        if x == 0:
            y += f
        elif f >= x:
            y += (f-x)
        else:
            y += f
    if y == 0:
        print("Đây là một mảng đẹp!")
    else:
        print(f"Đây không phải là một mảng đẹp. Cần xóa y = {y} số.")
    if __name__ == "__main__":
        solve()
```

```python
def find_majotity_element(nums):
    candidate = None
    count = 0
    for num in nums:
        if count == 0:
            candidate = num
            count = 1
        elif num == candidate:
            count += 1
        else:
            count -= 1
    if candidate is not None:
        actual_count = sum(1 for x in nums if x == candidate)
        if actual_count > len(nums) // 2:
            print(candidate)

```

```python
day_so = list(map(int, input().split()))
chan = []
le = []
for x in day_so:
    if x % 2 == 0:
        chan.append(x)
    else:
        le.append(x)
chan.sort()
le.sort(reverse=True)
tong = chan + le
print(tong)
```
