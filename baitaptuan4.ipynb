```python
m = int(input())
n = int(input())
uoc_m = []
uoc_n = []

for i in range(1, m+1):
    if m % i == 0 and i != m:
        uoc_m.append(i)

for i in range(1, n+1):
    if n % i == 0 and i != n:
        uoc_n.append(i)
tong_m = 0
tong_n = 0
for i in uoc_m:
    tong_m += i
for i in uoc_n:
    tong_n += i
if tong_m == n and tong_n == m:
    print("true")
else:
    print("false")

```

```python
import math
def kiem_tra_so_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
while True:
    try:
        n = int(input())
        if n > 0:
            break
        else:
            print("so vua nhap khong phai so nguyen to. Vui long nhap lai.")
    except ValueError:
        print("Du lieu nhap vao khong hop le. Vui long nhap lai.")
if kiem_tra_so_nguyen_to(n):
    print(f"{n} la so nguyen to.")
else:
    print(f"{n} khong phai la so nguyen to.")
```
