```python
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
    def getArea(self):
        return 2 * 3.14 * self.radius * self.height + 2 * 3.14 * self.radius ** 2
    def getVolume(self):
        return 3.14 * self.radius ** 2 * self.height
    def __str__(self):
        return f"({self.radius}, {self.height})"
b, c = map(float, input().split())
Cylinder1 = Cylinder(b, c)
print(Cylinder1)
print(f"Dien tich be mat: {Cylinder1.getArea()}")
print(f"The tich: {Cylinder1.getVolume()}")
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def goc(self):
        return self.x == 0 and self.y == 0
    def ox(self):
        return self.y == 0 and self.x != 0
    def oy(self):
        return self.x == 0 and self.y != 0
    def khoang_cach(self, other):
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5
m = list(map(float, input().split()))
p = Point(m[0], m[1])
if p.goc():
    print("diem goc toa do")
    
if p.ox():
    print("diem tren truc ox")
if p.oy():
    print("diem tren truc oy")
ket_qua = p.khoang_cach(Point(1, 2))
print(f"Khoang cach tu diem {p} den diem (1, 2) la: {ket_qua}")
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = []
    def add_item(self, item):
        self.items.append(item)
    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
    def get_total(self):
        return sum(self.items)
cart = ShoppingCart()
cart.add_item(10)
cart.add_item(20)
print(f"Total: {cart.get_total()}")
cart.remove_item(10)
print(f"Total sau khi xoa: {cart.get_total()}")

    
```
