```python
def move_zeros(arr):
    zeros = [x for x in arr if x == 0]
    non_zeros = [x for x in arr if x != 0]
    to_front = zeros + non_zeros
    to_back = non_zeros + zeros
    return to_front, to_back
input_arr = [0, 2, 3, 2, 4, 0, 0, 0, 9]
ket_qua_dau, ket_qua_cuoi = move_zeros(input_arr)
print(f"Đẩy_về_đầu: {ket_qua_dau}")
print(f"Đẩy_về_cuối: {ket_qua_cuoi}")
```
