```python
def la_so_amstronhg(x):
    s = str(x)
    p = len(s)
    tong = 0
    for i in s:
        tong += int(i) ** p
    if tong == x:
        return "la so amstronhg"
    else:
        return "khong la so amstronhg"
x = int(input())
print(la_so_amstronhg(x))
```

```python
def detect_number_type(s):
    if s.lstrip('-').isdigit():
        return 'int'
    if s .count('-') == 1:
        parts = s.split('.')
        if len(parts) == 2:
            left = parts[0].lstrip('-')
            right = parts[1]
            if left.isdigit() and right.isdigit():
                return 'float'
    return 'neither'
s = input()
result = detect_number_type(s)
print(result)
```

```python
def ba_diem(x1, y1, x2, y2, x3, y3):
    if (x1 - x2)/(x1 - x3) == (y1 - y2)/(y1 - y3):
        return "3 diem thang hang"
    else:
        return "3 diem khong thang hang"
x1, y1, x2, y2, x3, y3 = map(float, input().split())
print(ba_diem(x1, y1, x2, y2, x3, y3))
```

```python
def goc_giua(x1, y1, x2, y2, x3, y3):
    import math
    goc = ((x1 - x2) * (x3 - x2) + (y1 - y2) * (y3 - y2)) / (math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2) * math.sqrt((x3 - x2) ** 2 + (y3 - y2) ** 2))
    return math.cos(goc)
x1, y1, x2, y2, x3, y3 = map(float, input().split())
print(goc_giua(x1, y1, x2, y2, x3, y3))
```
