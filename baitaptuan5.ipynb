```python
import math

def so_nguyen_to(n):
    if n < 2:
        return "n khong la so nguyen to"
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return "n khong la so nguyen to"
    return "n la so nguyen to"
n = int(input("Enter a number: "))
print(so_nguyen_to(n))
```

```python
def so_hoan_hao(n):
    if n < 1:
        return "n khong la so hoan hao"
    uoc = []
    for i in range(1, n):
        if n % i == 0:
            uoc.append(i)
            tong = 0
            for i in uoc:
                tong += i
                if tong == n:
                    return "n la so hoan hao"
                else:
                    return "n khong la so hoan hao"
n = int(input("Enter a number: "))
print(so_hoan_hao(n))
    
```

```python
def tim_vi_tri(danh_sach, k):
    for i in range(len(danh_sach)):
        if danh_sach[i] == k:
            return i
        return -1
danh_sach = list(map(int, input("Enter a list of numbers: ").split()))
k = int(input("Enter a number to find: "))
print(tim_vi_tri(danh_sach, k))
```

```python
def giai_thua(n):
    if n < 0:
        return "n khong hop le"
    giai_thua = 1
    for i in range(1, n+1):
        giai_thua *= i
    return giai_thua
n = int(input())
print(giai_thua(n))
```
